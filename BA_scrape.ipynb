```python
title=[]
name=[]
country=[]
date_of_review=[]
passenger=[]
recommended=[]
rating_Value=[]
eco=[]
premium=[]
business=[]
first=[]
family=[]
couple=[]
solo=[]
busi=[]
year_22=[]
year_21=[]
year_20=[]
year_19=[]
year_18=[]
review=[]
```

```python
pages=np.arange(1,35)
```

```python
for page in pages:
    page=requests.get("https://www.airlinequality.com/airline-reviews/british-airways/page/"+str(page)+"/?sortby=post_date%3ADesc&pagesize=100")
    soup1=BeautifulSoup(page.content,'html.parser')
    data=soup1.find_all('article',attrs={'itemprop':'review'})
    sleep(randint(3,9))
    for store in data:
        title1=store.h2.text
        title.append(title1)
    
        name1=store.find('span',attrs={'itemprop':'name'}).text
        name.append(name1)
    
        country1=store.h3.text
        if('(' in country1):
            br = '('
            pos1=int(country1.index(br))
            br1 = ')'
            pos2=int(country1.index(br1))
            country1=country1[pos1+1:pos2]
        else:
            country1=""
        country.append(country1)
    
        d1=store.find('time',attrs={'itemprop':'datePublished'}).text
        date_of_review.append(d1)
        
        passenger1=store.strong.em.text if (store.find('strong') and store.find('em')) else ""
        passenger.append(passenger1)
        
        rec=store.find('td', class_ = "review-value rating-yes").text if store.find('td', class_ = "review-value rating-yes") else store.find('td', class_ = "review-value rating-no").text
        recommended.append(rec)
        
        rate=store.find('span',attrs={'itemprop':'ratingValue'}).text if store.find('span',attrs={'itemprop':'ratingValue'}) else ""
        rating_Value.append(rate)
        
        rev=store.find('div',class_ = "text_content").text if store.find('div', class_ = "text_content") else ""
        review.append(rev)
        
    string=str(data)
    count_economy=string.count('Economy Class')
    eco.append(count_economy)
    count_business=string.count('Business Class')
    business.append(count_business)
    count_premium=string.count('Premium Economy')
    premium.append(count_premium)
    count_first=string.count('First Class')
    first.append(count_first)
    count_family=string.count('Family Leisure')
    family.append(count_family)
    count_couple=string.count('Couple Leisure')
    couple.append(count_couple)
    count_solo=string.count('Solo Leisure')
    solo.append(count_solo)
    count_busi=string.count('Business')
    busi.append(count_busi)
    count_22=string.count('2022')
    year_22.append(count_22)
    count_21=string.count('2021')
    year_21.append(count_21)
    count_20=string.count('2020')
    year_20.append(count_20)
    count_19=string.count('2019')
    year_19.append(count_19)
    count_18=string.count('2018')
    year_18.append(count_18)
```

```python
count_demo=pd.DataFrame({"Economy Class":eco,"Business Class":business,"Premium Economy Class":premium,"First Class":first})
type_demo=pd.DataFrame({"Solo":solo,"Family":family,"Business":busi,"Couple":couple})
year_demo=pd.DataFrame({"Y 2022":year_22,"Y 2021":year_21,"Y 2020":year_20,"Y 2019":year_19,"Y 2018":year_18})
```
